```python
mask = stroke['gender'].isin(['Other'])
stroke = stroke[~mask]

stroke
```

```python
fig,axes = plt.subplots(4,2,figsize = (16,16))
fig.suptitle("Count plot for various categorical features")

sns.countplot(ax=axes[0,0],data=stroke,x='gender')
sns.countplot(ax=axes[0,1],data=stroke,x='hypertension')
sns.countplot(ax=axes[1,0],data=stroke,x='heart_disease')
sns.countplot(ax=axes[1,1],data=stroke,x='ever_married')
sns.countplot(ax=axes[2,0],data=stroke,x='work_type')
sns.countplot(ax=axes[2,1],data=stroke,x='Residence_type')
sns.countplot(ax=axes[3,0],data=stroke,x='smoking_status')
sns.countplot(ax=axes[3,1],data=stroke,x='stroke')

plt.show()
```

```python
fig,axes = plt.subplots(4,2,figsize = (16,16))
fig.suptitle("Count plot for various categorical features use hue")


ax1 = sns.countplot(ax = axes[0, 0], x = 'gender', hue = 'stroke', data = stroke)
ax2 = sns.countplot(ax = axes[0, 1], x = 'hypertension', hue = 'stroke', data = stroke)
ax3 = sns.countplot(ax = axes[1, 0], x = 'heart_disease', hue = 'stroke', data = stroke)
ax4 = sns.countplot(ax = axes[1, 1], x = 'ever_married', hue = 'stroke', data = stroke)
ax5 = sns.countplot(ax = axes[2, 0], x = 'work_type', hue = 'stroke', data = stroke)
ax6 = sns.countplot(ax = axes[2, 1], x = 'Residence_type', hue = 'stroke', data = stroke)
ax7 = sns.countplot(ax = axes[3, 0], x = 'smoking_status', hue = 'stroke', data = stroke)

for p in ax1.patches:
    height = p.get_height()
    ax1.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)

for p in ax2.patches:
    height = p.get_height()
    ax2.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)
    
for p in ax3.patches:
    height = p.get_height()
    ax3.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)

for p in ax4.patches:
    height = p.get_height()
    ax4.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)
    
for p in ax5.patches:
    height = p.get_height()
    ax5.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)
    
for p in ax6.patches:
    height = p.get_height()
    ax6.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)
    
for p in ax7.patches:
    height = p.get_height()
    ax7.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)
```
